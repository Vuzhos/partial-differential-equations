```python
import numpy as np
from math import sqrt
%matplotlib widget
import matplotlib.pyplot as plt
from matplotlib import animation, rc
```

```python
def mu2(x):
    return [1/(1 + ((t-50)/10)**4) for t in x]
```

```python
h = 1
tau = 0.05
x = np.arange(0, 100 + h, h)
t = np.arange(0, 1000 * tau + tau, tau)

u = np.zeros((len(t), len(x)))
u[0] = mu2(x)
for i in range(len(t)):
    u[i][0] = u[0][0]

for i in range(1, len(t)):
    for j in range(1, len(x)):
        u[i][j] = (-h + sqrt(h**2 + tau * (2 * h * u[i-1][j] + tau * u[i-1][j-1] ** 2))) / tau
```

```python
fig, ax = plt.subplots()
ax.grid(True)
line, = plt.plot(x, u[0], color='r')
a = animation.FuncAnimation(fig, animate, frames=len(t), interval=1)
plt.show()
```
